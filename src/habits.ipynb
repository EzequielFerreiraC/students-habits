```python
# importando as bibliotecas

import pandas
import seaborn
import matplotlib.pyplot as plt
```

```python

# chamando o pandas para ler o meu csv
dataFrame = pandas.read_csv(r"../data/student_habits_performance.csv")

# Mostra as informações da tabela
# dataFrame.info()

cols = [
    "study_hours_per_day",
    "social_media_hours",
    "netflix_hours",
    "sleep_hours",
    "attendance_percentage",
    "exercise_frequency",
    "mental_health_rating",
    "exam_score"
]

# Plotando o mapa, calculando a correlação, com o corr()
# Dois parâmetros: annot=True (anotar os valores da correlação), coolwarm mudar as cores
seaborn.heatmap(dataFrame[cols].corr(), annot=True, cmap="coolwarm")
```

```python
# Gráfico de dispersão com linha de regrassão

seaborn.lmplot(data=dataFrame, x="study_hours_per_day", y="exam_score")
plt.title("Mais estudo -> notas mais altas?")

plt.show()
```

```python
filtroEstudoAlto = dataFrame["study_hours_per_day"] > 5
filtroEstudoBaixo = dataFrame["study_hours_per_day"] > 2

grupoEstudoAlto = dataFrame[filtroEstudoAlto]["exam_score"]
grupoEstudoBaixo = dataFrame[filtroEstudoBaixo]["exam_score"]


print("Média notas (estuda > 5h):", round(grupoEstudoAlto.mean(), 2))
print("Média notas (estuda < 2h):", round(grupoEstudoBaixo.mean(), 2))
```

```python
seaborn.histplot(data=dataFrame, x="social_media_hours")
plt.title("Distribuição das Redes Sociais")
plt.show()
```
